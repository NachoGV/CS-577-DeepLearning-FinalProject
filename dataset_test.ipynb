## Imports

```python
# Imports
import os
import cv2 as cv
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Requirements 
!pip freeze > requirements.txt
```

## Dataset

```python
# Get paths and labels
base_path = 'ExDark/'
df = pd.DataFrame(columns=['Path', 'Image', 'Label'])
df['Path'] = [os.path.join(base_path, label, image) for label in os.listdir(base_path) for image in os.listdir(os.path.join(base_path, label))]
df['Label'] = [os.path.basename(os.path.dirname(path)) for path in df['Path']]
df['Image'] = [cv.imread(path) for path in df['Path']]
df = df.sample(frac=1).reset_index(drop=True) # Shuffle Dataset
df.head(10)
```

```python
# OHE Labels
df = pd.get_dummies(df, columns=['Label'], dtype=np.int8)
df.head(10)
```
